# Sentence Tokenization

```python
from nltk.tokenize import sent_tokenize
```

```python
import nltk
nltk.download('punkt_tab')
```

```python
text = "Hello, My name is Rudra Kashyap. How are you doing?"
sent_tokenize(text)
```

```python
tokenizer = nltk.data.load('tokenizers/punkt/PY3/english.pickle') # used for large data as this tokenizer trained on large corpus
tokenizer.tokenize(text)
```

# Word Tokenization

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(text)
```

```python
from nltk.tokenize import WordPunctTokenizer
```

```python
tokenizer = WordPunctTokenizer()
tokenizer.tokenize(text)
```

```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'\w+')
tokenizer.tokenize(text)
```
